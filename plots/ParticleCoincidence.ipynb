# calculate particle coincidence
in this notebook the probability for particle coincidence is being calculated for the GFAS sampling. Adapted from IGOR code written by Darrel Baumgardner \
Probability of particle coincidence: the probability that the space between two arriving particles is smaller than the laser beam width. They are therefore measured at the same time.

author: Lea Haberstock, Stockholm University, Department of Environmental Science, Atmospheric Unit 

based on: \
Baumgardner, Strapp, Dye (1985): Evaluation of the Forward Scattering Spectrometer Probe. Part II: Corrections for coincidence and dead-time losses \

```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib 
```

```python
concentration = np.logspace(0,4, 500)
avgW =  concentration**(-1/3)
beamwidth = 220 # for the GFAS and in µm 
```

```python
# probabilty of coincidence 
P_c = 100*(1- np.exp(-(beamwidth /10000)/avgW)) # in %, the 10000 is to convert from µm to cm, which is the unit of avgW
```

```python
# figsize 5,4
plt.figure(figsize=(5,3))
plt.plot(concentration, P_c, color = 'k' )
plt.xlabel('Number concentration (cm$^{-3}$)', fontsize = 12)
plt.ylabel('P$_c$ (%)', fontsize = 12)
plt.ylim(1,40)
plt.xlim(1,10000)
plt.xscale('log')
plt.grid()
plt.savefig('coincidence_probability.png', dpi = 300, bbox_inches = 'tight')
```
